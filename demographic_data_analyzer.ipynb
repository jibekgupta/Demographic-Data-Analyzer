### What is the average age of men?


```python
men = df.loc[df['sex']=='Male']
avg_men_age = men.age.mean()
float(round(avg_men_age,1))
```

### What is the percentage of people who have a Bachelor's degree?


```python
edu_bachelors = df.loc[df['education']=='Bachelors']
total_people = len(df)
percentage_bachelors = ((len(edu_bachelors) / total_people) * 100)
round(percentage_bachelors, 1)
```

### What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?


```python
num_advanced_edu = df.loc[df['education'].isin(['Bachelors','Masters', 'Doctorate'])]
len(num_advanced_edu)
```

```python
num_advanced_edu_rich = num_advanced_edu.loc[num_advanced_edu['salary']=='>50K']
len(num_advanced_edu_rich)
```

```python
percentage_advanced_edu_rich = (len(num_advanced_edu_rich) / len(num_advanced_edu)) * 100
round(percentage_advanced_edu_rich, 1)
```

### What percentage of people without advanced education make more than 50K?


```python
edu_lower = df.loc[~df['education'].isin(['Bachelors','Masters', 'Doctorate'])]
len(edu_lower)
```

```python
edu_lower_rich = edu_lower.loc[edu_lower['salary']=='>50K']
len(edu_lower_rich)
```

```python

percentage_edu_lower_rich = (len(edu_lower_rich) / len(edu_lower)) * 100
round(percentage_edu_lower_rich, 1)
```

### What is the minimum number of hours a person works per week?

```python
df.columns
```

```python
min_work_hours =df['hours-per-week'].min()
min_work_hours
```

### What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?


```python
min_hour_workers = df[df['hours-per-week']==min_work_hours]
len(min_hour_workers)
```

```python
min_hour_high_salary = min_hour_workers[min_hour_workers['salary']=='>50K']
rich_percantage = (len(min_hour_high_salary) / len(min_hour_workers)) *100
round(rich_percantage,1)
```

## What country has the highest percentage of people that earn >50K and what is that percentage?

```python
country_counts = df['native-country'].value_counts()
len(country_counts)
```

```python
rich_country_counts = df.loc[df['salary']=='>50K']['native-country'].value_counts()
len(rich_country_counts)
```

```python
percentage_rich_by_country = rich_country_counts / country_counts * 100
highest_earning_country = percentage_rich_by_country.idxmax()
highest_earning_country
```

```python
highest_earning_country_percentage = round(percentage_rich_by_country.max(),1)
highest_earning_country_percentage
```
